```python
import matplotlib.pyplot as plt
```

```python
bank['job'].value_counts().plot(kind='pie',autopct='%.2f',legend=True)
plt.legend(bbox_to_anchor=(1.5,1.0))
```

```python
bank['age'].plot(kind='hist',bins=20,legend=True)
plt.title('Age')
plt.xlabel('Height')
```
